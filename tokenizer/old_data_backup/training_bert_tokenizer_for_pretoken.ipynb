```python

from __future__ import print_function
from collections import defaultdict
import os
import sys
import posixpath

from elftools.elf.elffile import ELFFile

import os
import json 

import sys,os

```

```python


special_tokens = ["[address]", "[mask]"] 
```

```python
DATA_PATH = "/home/raisul/ANALYSED_DATA/tokenization_data_single_functions"

json_files = [os.path.join(DATA_PATH, f) for f in os.listdir(DATA_PATH) ]

print(len(json_files))
print(json_files[0])
original = []
def read_corpus():

    all = []

    for k, j_file in enumerate(json_files):
        # if k>20:
        #     break
        try:
            with open(j_file, 'r') as file:
                funct_data = json.load( file )

                
                all.extend( [funct_data['disassembly_decimal'] , 
                             funct_data ['disassembly_all_number_to_words'] ,
                             funct_data ['disassembly_addresses_to_words']])
                original.append(  funct_data['disassembly_original'])
                original.append(  funct_data['disassembly_original'])
                original.append(  funct_data['disassembly_original'])
        except Exception as e:
            # Print the error message
            print(f"An error occurred: {e}")
    return all
    


training_corpus = read_corpus()



print("Functions Count: ",len(training_corpus), '\n')
example = training_corpus[10]


```

```python
from transformers import AutoTokenizer

old_tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
old_tokenizer.pad_token
```

```python

tokenizer = old_tokenizer.train_new_from_iterator(training_corpus, vocab_size=200000, min_frequency=1 , additional_special_tokens = special_tokens )
special_tokens = {"additional_special_tokens": ["\n" , ' ']}
tokenizer.add_special_tokens(special_tokens)
```

```python

for i in range(100):
    tokens = tokenizer.tokenize(training_corpus[i])
    print(original[i])
    print(i%3,''.join(tokens))
```
